```python
files = os.listdir("Sales_Data")
files
```

## Read the files !

```python
df = pd.DataFrame()
```

```python
for i in files:
    data = pd.read_csv(os.path.join("Sales_Data",i))
    df = pd.concat([df,data])
df
```

```python
df.dropna(inplace = True)
df
```

```python
filter_text = df['Quantity Ordered'] != 'Quantity Ordered'

df = df[filter_text]
```

```python
df['Quantity Ordered'] = df['Quantity Ordered'].astype('float')
df['Price Each'] = df['Price Each'].astype('float')
df.info()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["year"] = df["Order Date"].dt.year
df["month"] = df["Order Date"].dt.month
df["hour"] = df["Order Date"].dt.hour
df["minute"] = df["Order Date"].dt.minute
df
```

```python
df
```

```python
# Q1
df["Revenue"] = df["Quantity Ordered"] * df["Price Each"]
df
```

```python
df.groupby("year").agg({"Revenue" : "sum"})
```

```python
 #Q2
df.groupby("month").agg({"Revenue":"mean"})
df
```

```python
#Q3
df.groupby("month").agg({"Revenue":"mean"}).sort_values("Revenue",ascending = False)
df
```

```python
#Q4
city = []
for i in df["Purchase Address"]:
    c = i.split(",")[1]
    city.append(c)
df["city"] = city
df
```

```python
df.groupby("city").agg({"Revenue" : "sum"}).sort_values("Revenue" , ascending = False)
```

```python
#Q5
df.groupby("Product").agg({"Quantity Ordered" : "sum"}).sort_values("Quantity Ordered" , ascending = False)
```

```python
#Q6
df.groupby("hour").agg({"Quantity Ordered" : "sum"}).sort_values("Quantity Ordered" , ascending = False)
```
